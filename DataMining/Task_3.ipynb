```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

df = pd.read_csv('shoesize.csv')

female_data = df[df['Gender'] == 'F']
male_data = df[df['Gender'] == 'M']

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(data=female_data, x='Size', y='Height')
plt.title('Size vs Height (Females)')
plt.xlabel('Size')
plt.ylabel('Height')

plt.subplot(1, 2, 2)
sns.scatterplot(data=male_data, x='Size', y='Height')
plt.title('Size vs Height (Males)')
plt.xlabel('Size')
plt.ylabel('Height')

plt.tight_layout()
plt.show()

female_corr, _ = pearsonr(female_data['Size'], female_data['Height'])
print(f"Pearson's correlation for females: {female_corr:.2f}")

male_corr, _ = pearsonr(male_data['Size'], male_data['Height'])
print(f"Pearson's correlation for males: {male_corr:.2f}")
```
